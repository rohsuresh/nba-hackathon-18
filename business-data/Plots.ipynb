```python
import pandas as pd

game_data = pd.read_csv('./game_data.csv')
player_data = pd.read_csv('./player_data.csv')
training_set = pd.read_csv('./training_set.csv')
```

```python
game_data
```

```python
player_data.head()
```

```python
game_ids = list(sorted(set(game_data.Game_ID)))

stars_in_games = []
for gid in game_ids:
    game = player_data.loc[player_data['Game_ID'] == gid]
    all_star_status = list(game.ASG_Team)
    num_all_stars = sum([1 if player != 'None' else 0 for player in all_star_status])
    stars_in_games.append(num_all_stars)

d = {'Game_ID': game_ids, 'NumAllStars': stars_in_games}
all_stars_df = pd.DataFrame(data=d)
all_stars_df.head()
```

```python
list(training_set.Game_ID)
```

```python
viewers = training_set.groupby(['Game_ID']).sum()
viewers.head()
```

```python
stars_and_viewers = all_stars_df.join(viewers, on='Game_ID')
stars_and_viewers.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(stars_and_viewers.NumAllStars, stars_and_viewers['Rounded Viewers'])
plt.show()
#stars_and_viewers.plot('NumAllStars', 'Rounded Viewers')
```
